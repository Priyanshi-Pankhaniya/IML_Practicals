```python
'''
B. to round elements of the array to the nearest integer
'''

#IMporting the Numpy
import numpy as np

#Printing the Round of 3.3
print("Round of 3.3 is:",np.round(3.3));
```

```python
'''
C. to calculate mean across dimension, in a 2D numpy array
'''

#Importing the numpy
import numpy as np

#Creating an 2-D array
a=np.array([[1,2,3],[2,5,7],[4,6,7]]);

#Printing the Mean
print("Mean:",np.mean(a));
```

```python
'''
D. to calculate the difference between neighboring elements, element-wise of a given array
'''

#Importing Numpy
import numpy as np

#Creating array
a=np.array([[45,23,78],[23,45,6]]);

#Calcuting and printing the difference between the neighbouring elements
print("The difference between neighbouring elements is:",np.diff(a,axis=1));
print(np.diff(a, axis=0));
```
