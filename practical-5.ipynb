```python
'''
B. to compute the mean, standard deviation, and variance of a given array along the second 
axis
'''

#Importing the numpy
import numpy as np

#Creating the array
a=np.array([[12,13,14,34],[23,34,45,56]]);

#Calculating the mean
print("Mean:",np.mean(a,axis=1));

#Calculating the deviation
print("Deviation",np.std(a,axis=1));

#Calculating the variance
print("Variance:",np.var(a,axis=1));
```
