```python
'''
Enroll no.: 216270307066
Name: Priyanshi Pankhaniya
Div:G
Batch: G3
Write a Pandas program to create a line plot of the opening, closing stock
prices of given company between two specific dates. 
'''
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/kaggle/input/nifty50-stock-market-data/ICICIBANK.csv');
limited_df=df.iloc[:130]
date_range=pd.date_range("2000-03-07",periods=len(limited_df),freq="D")
                 
limited_df.set_index(date_range,inplace=True)


# Create a line plot for opening and closing prices
plt.figure(figsize=(12, 6))
plt.plot(limited_df.index, limited_df['Open'], label='Open',color='green')
plt.plot(limited_df.index, limited_df['Close'], label='Close',color='red')
plt.title('Opening and Closing Stock Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
