```python
'''
B. to create a 3x3 matrix with values ranging from 2 to 10
'''

#Importing Numpy
import numpy as np

#Converting it into 3x3 matrix
a=np.arange(2,11).reshape(3,3);

#Printing the shape
print("Shape:",a.shape);

#Printing the Matrix
print("3x3 matrix:",a);
```

```python
'''
C. to append values at the end of an array
'''
#Importing Numpy
import numpy as np

#Creating Array
arr=np.array(["Jeannie","Jisoo"]);

print("Before Appending:",arr);

#Creating Array
a=np.append(arr,["Rose","Lisa"]);

#Printing the array after appending the elements
print("After appending:",a);
```

```python
'''
D. to create another shape from an array without changing its data(3*2
to 2*3)
'''

#Importing Numpy
import numpy as np

#Creating a array of 3x2
arr=np.arange(2,8).reshape(3,2);

#Printing the array
print("3x2 matrix:",arr);

#Reshaping a 3x2 to 2x3
arr1=arr.reshape(2,3);

#Printing the Array
print("2x3 matrix:",arr1);
```
